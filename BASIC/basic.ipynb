```python
# both the above does the same work but one used apply function and one broadcasting , broadcasting is useful for simple operations while apply 
# for complex stuffs . 
# broadcasting is only limited to + , -  , * , ? ,% ,> , < , == , >= , <= ,etc you can't use if else and all things . 
# So simple examples for all 
# 1, when i want to filter the data like only a data wheere P, Min Mass(EU) >1000 then use normal filtering 
# 2. when i want to muliply each row of the Min value by 2 then use broadcastsing 
#3. when i want to do complex operations on each row then use use apply
# instead of lambda you can make a separate functions  ,# it is usually necessary to return some value for apply otherwise it is null
```
